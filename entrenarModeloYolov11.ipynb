```python
from ultralytics import YOLO
import os
```

```python
# =============================================================================================
# copiar la carpeta comprimida con el dataset en formato Yolo en /content/ y luego descomprimir
# =============================================================================================
!unzip -q /content/data.zip -d /content/
```

```python
# Cargar el modelo preentrenado
model = YOLO('yolo11s.pt')  # Puedes cambiar a yolo11s.pt, yolo11m.pt, etc.
```

```python
# Entrenar el modelo
model.train(
    data='/content/data.yaml',  # Ruta al archivo .yaml
    epochs=10,  # N√∫mero de √©pocas (ajusta seg√∫n tu dataset)
    imgsz=640,  # Tama√±o de las im√°genes (aj√∫state a tu dataset)
    batch=32,   # Tama√±o del lote (ajusta seg√∫n la memoria de la GPU)
    optimizer = 'SGD',
    name='yolo11l_datasetv10_10epocas'  # Nombre del experimento
)
```

```python
# Evaluar el modelo
results = model.val()
print(results)
```

```python
# Hacer predicciones en una imagen
results = model.predict(source='/content/nordelta03.png', save=True)
```
